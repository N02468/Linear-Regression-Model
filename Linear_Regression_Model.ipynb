**Import Linear Regression with Single Variable**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("homeprices.csv.xlsx")
print(df)
```

```python
df.drop(columns=['Bill','Gas'],axis='columns',inplace=True)
df
```

```python
print(df)
```

```python
model = linear_model.LinearRegression()
model.fit(df[['Area']],df['Price'])
```

```python
plt.plot(df.Area,df.Price)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression')
plt.show()
```

```python
model.predict([[3400]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
m=model.coef_
x=3400
c=model.intercept_
y=m*x+c
y
```
